# GridSearchCV

```python
data = pd.read_csv('https://raw.githubusercontent.com/ingledarshan/BK_Birla/main/Social_Network_Ads.csv')
data.shape
```

```python
data.head()
```

```python
X = data.iloc[ : , 1:4]
X.shape
```

```python
X.head()
```

```python
X.Gender.replace({'Male':0, 'Female':1}, inplace=True)
X.head()
```

```python
y = data.Purchased
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
X_train[['Age','EstimatedSalary']].head(2)
```

```python
X_train[['Age','EstimatedSalary']] = mms.fit_transform(X_train[['Age','EstimatedSalary']])
X_train[['Age','EstimatedSalary']].head()
```

```python
X_test[['Age','EstimatedSalary']] = mms.transform(X_test[['Age','EstimatedSalary']])
X_test[['Age','EstimatedSalary']].head()
```

```python
X_train.head()
```

```python
classifyer = DecisionTreeClassifier()
```

```python
classifyer.fit(X_train, y_train)
```

```python
plot_tree(classifyer);
```

```python
y_pred_train = classifyer.predict(X_train)
y_pred_test = classifyer.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train, y_pred_train)
```

```python
accuracy_score(y_test, y_pred_test)
```

```python
param_dist = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : range(1,10),
    'min_samples_leaf' : range(1,20),
    'max_features' : ["auto", "sqrt", "log2"],
    'max_leaf_nodes' : [1,2,3,4,5]
}
param_dist
```

```python
from sklearn.model_selection import GridSearchCV
grid1 = GridSearchCV(classifyer, param_grid=param_dist, cv=10, n_jobs=-1)

# http://ethen8181.github.io/machine-learning/model_selection/img/kfolds.png
```

```python
grid1.fit(X_train, y_train)
```

```python
grid1.best_params_
```

```python
grid1.best_estimator_
```

```python
grid1.best_score_
# https://www.analyticsvidhya.com/blog/2021/06/tune-hyperparameters-with-gridsearchcv/
```

```python
param_dist = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [1,2,3,4,5,6,7,None]
}
param_dist
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(classifyer, param_grid=param_dist, cv=10, n_jobs=-1)

# http://ethen8181.github.io/machine-learning/model_selection/img/kfolds.png
```
